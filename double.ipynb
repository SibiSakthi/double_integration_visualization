```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import dblquad
import sympy as sp

# Define the function to be integrated
def integrand(y, x):
    return 1

# Define the limits of the integration
x_min, x_max = 1/2, 1  # Example limits for x
y_min_func = lambda x: 1/(2*x)
y_max_func = lambda x: 1/x  # Example limit for y

# Convert the functions to symbolic expressions
x_sym = sp.Symbol('x')
y_min_expr = 1/(2*x_sym)
y_max_expr = 1/x_sym

# Convert the symbolic expressions to strings
y_min_str = sp.simplify(y_min_expr)
y_max_str = sp.simplify(y_max_expr)

# Compute the double integral
result, error = dblquad(integrand, x_min, x_max, y_min_func, y_max_func)

print(f"Double integral result: {result}, with error: {error}")

# Create a figure for the plot
plt.figure(figsize=(10, 6))

# Define x values for plotting the boundaries
x_vals = np.linspace(x_min, x_max, 500)
y_min_values = y_min_func(x_vals)
y_max_values = y_max_func(x_vals)

# Highlight the area of integration using fill_between
plt.fill_between(x_vals, y_min_values, y_max_values, color='red', alpha=0.3, edgecolor='red')

# Plot the lower boundary function y = 1/(2x)
plt.plot(x_vals, y_min_values, label=f'y = {y_min_str}', color='blue')

# Plot the upper boundary y = 1/x
plt.plot(x_vals, y_max_values, label=f'y = {y_max_str}', color='green')

# Plot the vertical boundaries x = 1/2 and x = 1
plt.axvline(x=x_min, color='purple', linestyle='--', label=f'x = {x_min}')
plt.axvline(x=x_max, color='orange', linestyle='--', label=f'x = {x_max}')

# Add plot details
plt.title('Highlighted Area of Integration with Boundary Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_min - 0.5, x_max + 0.5)
plt.ylim(0, 2)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import dblquad

# Define the function to be integrated
def integrand(y, x):
    return 1

# Define the limits of the integration
x_min_func = lambda y: 1 / (2 * y)  # Function for lower limit of x
x_max_func = lambda y: 1  # Constant upper limit for x
y_min = 0.5  # Constant lower limit for y
y_max = 1  # Constant upper limit for y

# Compute the double integral
result, error = dblquad(integrand, y_min, y_max, lambda y: x_min_func(y), lambda y: x_max_func(y))

print(f"Double integral result: {result}, with error: {error}")

# Create a figure for the plot
plt.figure(figsize=(10, 6))

# Define y values for plotting the boundaries
y_vals = np.linspace(y_min, y_max, 500)
x_min_values = x_min_func(y_vals)
x_max_values = np.full_like(y_vals, x_max_func(y_vals[0]))

# Highlight the area of integration using fill_between
plt.fill_betweenx(y_vals, x_min_values, x_max_values, color='red', alpha=0.3, edgecolor='red')

# Plot the boundary functions
plt.plot(x_min_values, y_vals, label=r'$x = \frac{1}{2y}$', color='blue')
plt.plot(x_max_values, y_vals, label='x = 1', color='green')

# Plot the horizontal boundaries y = 0.5 and y = 1
plt.axhline(y=y_min, color='purple', linestyle='--', label=f'y = {y_min}')
plt.axhline(y=y_max, color='orange', linestyle='--', label=f'y = {y_max}')

# Add plot details
plt.title('Highlighted Area of Integration with Boundary Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 2)
plt.ylim(y_min - 0.1, y_max + 0.1)
plt.grid(True)
plt.legend()
plt.show()
```
